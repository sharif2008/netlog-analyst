```python
#Import & global configuration

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import (
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
# Optional gradient boosting libraries
try:
    from xgboost import XGBClassifier
except ImportError:
    XGBClassifier = None

try:
    from lightgbm import LGBMClassifier
except ImportError:
    LGBMClassifier = None

try:
    from catboost import CatBoostClassifier
except ImportError:
    CatBoostClassifier = None
    
import matplotlib.pyplot as plt

# Display options
pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 200)
```

## 2. Load dataset

Requirements:
- `Attack_Binary` column (0 = BENIGN, 1 = ATTACK).
- Remove NaNs, infinities, and duplicates
- Normalize the label into a binary target: BENIGN vs ATTACK.

```python
# TODO: adjust these paths to your local data directory
files = [
    "../data/Monday-WorkingHours.pcap_ISCX.csv",
    "../data/Wednesday-workingHours.pcap_ISCX.csv",
    "../data/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv",
    "../data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv",
]
def load_cic_file(path: str) -> pd.DataFrame:
    """Load a CIC-IDS-2017 CSV and normalize column names."""
    df = pd.read_csv(path)
    # strip spaces from column names
    df.columns = df.columns.str.strip()
    return df

dfs = []
for f in files:
    print("Loading:", f)
    df_tmp = load_cic_file(f)
    print("  Shape:", df_tmp.shape)
    dfs.append(df_tmp)
len(dfs)

# Find common columns across all dataframes
common_cols = set(dfs[0].columns)
for d in dfs[1:]:
    common_cols = common_cols.intersection(set(d.columns))

common_cols = sorted(list(common_cols))
print("Number of common columns:", len(common_cols))
print("Sample common columns:", common_cols[:20])

# Keep only common columns and concatenate
dfs_common = [d[common_cols].copy() for d in dfs]
df_full = pd.concat(dfs_common, axis=0, ignore_index=True)
print("Merged shape:", df_full.shape)
df_full.head()

df = df_full.copy()

# replace infinities with NaN
df = df.replace([np.inf, -np.inf], np.nan)

# drop rows with NaN values
df = df.dropna()

# drop duplicate rows
df = df.drop_duplicates()
df.head()
print("Original shape:", df.shape)
```

## 3. Stratified sampling (10% per class)

To speed up experiments, I have takne **10% from each class** (BENIGN and ATTACK). This keeps class distribution consistent.

```python
# Stratified 10% sampling
# inspect label values
df['Label'] = df['Label'].astype(str).str.strip()
df['Label'].value_counts()

# map to binary: 0 = BENIGN, 1 = any ATTACK
def map_binary(label: str) -> int:
    label = str(label).upper()
    if "BENIGN" in label:
        return 0
    else:
        return 1

df['Attack_Binary'] = df['Label'].apply(map_binary)
print (df['Attack_Binary'].value_counts())

df = df.groupby("Attack_Binary", group_keys=False).sample(frac=.1, random_state=52)

print("After stratified 10% sampling:", df.shape)
print(df["Attack_Binary"].value_counts())
print("Class distribution (%):")
print(df["Attack_Binary"].value_counts(normalize=True) * 100)
```

## 4. Correlation-based feature filtering

We remove highly correlated numeric features (|r| > 0.95) to reduce multicollinearity and redundancy.

- Only numeric columns are considered.
- `Attack_Binary` is excluded from being dropped.

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Exclude target from correlation-based filtering
corr_cols = [c for c in numeric_cols if c != "Attack_Binary"]

corr = df[corr_cols].corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

high_corr_threshold = 0.95
to_drop = [col for col in upper.columns if any(upper[col] > high_corr_threshold)]

print("Number of numeric features before correlation filter:", len(corr_cols))
print("Number of features flagged as highly correlated:", len(to_drop))

df_filtered = df.drop(columns=to_drop)
```

## 5. Define feature matrix `X` and target vector `y`

- `Label` and `Attack_Binary` from the feature set is removed.
- All remaining columns are used as features.

```python
# Final feature columns (exclude label/target)
filtered_feature_cols = [
    c for c in df_filtered.columns
    if c not in ["Label", "Attack_Binary"]
]

print("Final feature columns used for ML:")
print(filtered_feature_cols)
print("Number of final features:", len(filtered_feature_cols))

X = df_filtered[filtered_feature_cols].values
y = df_filtered["Attack_Binary"].values

print("X shape:", X.shape, " y shape:", y.shape)
```

## 5. Train/Test split (stratified)

We perform a 70/30 train/test split with stratification to maintain class distribution in both sets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape, " Test shape:", X_test.shape)
```

## 6. Scaling

We standardize numeric features using `StandardScaler`.

- Fit the scaler **only on training data**.
- Transform both train and test sets.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaled train shape:", X_train_scaled.shape)
```

## 7. Feature selection with RFE (RandomForest)

- Use a RandomForest as the base estimator.
- Select top 20 features using Recursive Feature Elimination.
- Fit RFE only on the **scaled training data**.

```python
rf_base = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rfe = RFE(
    estimator=rf_base,
    n_features_to_select=20,
    step=1
)

rfe.fit(X_train_scaled, y_train)

selected_mask = rfe.support_
selected_features = [f for f, m in zip(filtered_feature_cols, selected_mask) if m]

print("Selected features by RFE (top 20):")
for f in selected_features:
    print(" -", f)

X_train_sel = X_train_scaled[:, selected_mask]
X_test_sel = X_test_scaled[:, selected_mask]

print("Shape after RFE -> train:", X_train_sel.shape, " test:", X_test_sel.shape)
```

## 8. Define the models

We will train and compare the following algorithms:

- Logistic Regression  
- K-Nearest Neighbors
- Support Vector Machine
- Naive Bayes
- Decision Tree  
- Random Forest  
- Gradient Boosting  
- XGBoost (if installed)  
- LightGBM (if installed)  
- AdaBoost  
- CatBoost (if installed) 

```python
models = {}

# Classical ML
models["LogisticRegression"] = LogisticRegression(max_iter=1000, n_jobs=-1)
models["KNN"] = KNeighborsClassifier(n_neighbors=5)
models["SVM_RBF"] = SVC(kernel="rbf", probability=False, random_state=42)
models["NaiveBayes"] = GaussianNB()
models["DecisionTree"] = DecisionTreeClassifier(random_state=42)

# Ensemble tree models
models["RandomForest"] = RandomForestClassifier(
    n_estimators=200, random_state=42, n_jobs=-1
)
models["GradientBoosting"] = GradientBoostingClassifier(random_state=42)
models["AdaBoost"] = AdaBoostClassifier(random_state=42)

# Try importing XGBoost
try:
    from xgboost import XGBClassifier
    models["XGBoost"] = XGBClassifier(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric='logloss',
        random_state=42
    )
except ImportError:
    print("XGBoost not installed, skipping...")

# Try importing LightGBM
try:
    from lightgbm import LGBMClassifier
    models["LightGBM"] = LGBMClassifier(
        n_estimators=200,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
except ImportError:
    print("LightGBM not installed, skipping...")

# Try importing CatBoost
try:
    from catboost import CatBoostClassifier
    models["CatBoost"] = CatBoostClassifier(
        iterations=300,
        depth=6,
        learning_rate=0.1,
        verbose=0,
        random_state=42
    )
except ImportError:
    print("CatBoost not installed, skipping...")

models
```

## 9. Train & evaluate all models

For each model we compute:
- Accuracy
- Precision (weighted)
- Recall (weighted)
- F1-score (weighted)

We also print the full classification report for detailed per-class metrics.

```python
results = {}

for name, model in models.items():
    print(f"\n=== Training {name} ===")
    model.fit(X_train_sel, y_train)
    y_pred = model.predict(X_test_sel)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)

    print(classification_report(y_test, y_pred, target_names=["BENIGN", "ATTACK"], digits=5))

    results[name] = {
        "Accuracy (%)": acc * 100,
        "Precision (%)": prec * 100,
        "Recall (%)": rec * 100,
        "F1-Score (%)": f1 * 100
    }

results_df = pd.DataFrame(results).T.reset_index().rename(columns={"index": "Model"})
results_df = results_df.sort_values(by="Accuracy (%)", ascending=False)

print("\n=== Model Comparison (Test Set) ===")
results_df
```

## 10. Feature importance (RandomForest)

Feature importance using the RandomForest model trained on the RFE-selected feature subset.

```python
# Use the fitted RandomForest model
rf_final = models["RandomForest"]
importances = rf_final.feature_importances_

fi_df = pd.DataFrame({
    "Feature": selected_features,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print("=== Feature Importance (RandomForest on RFE-selected features) ===")
fi_df
```

### Plot feature importance

```python
plt.figure(figsize=(10, 6))
plt.barh(fi_df["Feature"], fi_df["Importance"])
plt.gca().invert_yaxis()  # highest importance at top
plt.xlabel("Importance")
plt.title("Feature Importance (RandomForest, RFE-selected features)")
plt.tight_layout()
plt.show()
```

## 11. Random flow prediction demo

We pick one random flow, run it through the same preprocessing pipeline, and compare the predicted vs actual label using the RandomForest model.

```python
# Sample one random row from the filtered dataframe
sample = df_filtered.sample(1, random_state=99)
sample_features = sample[filtered_feature_cols]

# Scale using the same scaler
sample_scaled = scaler.transform(sample_features)

# Apply RFE mask
sample_selected = sample_scaled[:, selected_mask]

# Predict with RandomForest
pred = rf_final.predict(sample_selected)[0]
actual = sample["Attack_Binary"].values[0]

print("=== Random Flow Prediction Demo (RandomForest) ===")
print("Actual Label   :", "ATTACK" if actual == 1 else "BENIGN")
print("Predicted Label:", "ATTACK" if pred == 1 else "BENIGN")
print("\nRaw features for this flow:")
sample[filtered_feature_cols]
```
